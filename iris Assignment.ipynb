```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x="petal_length",y="species",data=iris)
```

```python
sns.lineplot(x="petal_length",y="petal_width",data=iris)
```

```python
sns.distplot(iris['petal_length'])
```

```python
sns.barplot(x="sepal_length",y="sepal_width",data=iris)
```

```python
sns.boxplot(x="sepal_length",y="sepal_width",data=iris)
```

```python
sns.catplot(x="species",y="sepal_length",data=iris)
```

```python
sns.pairplot(iris)
```

```python
sns.heatmap(iris.corr(),annot=true)
```
